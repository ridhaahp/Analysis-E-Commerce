#E-Commerce Data Analysis

In the realm of e-commerce, data analysis is an integral process that involves the examination and interpretation of various datasets to derive valuable insights. The datasets typically comprise three key components: Customer Data, Order Data, and Reviews. Customer Data includes essential information such as customer ID and location, enabling businesses to understand the demographics and geographic distribution of their customer base. Order Data encompasses details like order names and category IDs, providing a comprehensive overview of the products or services that customers are purchasing. Lastly, the Review dataset incorporates crucial feedback in the form of review scores and comments, offering businesses the opportunity to gauge customer satisfaction and identify areas for improvement. Through the analysis of these datasets, e-commerce entities can optimize their strategies, personalize customer experiences, and enhance overall operational efficiency. The synergy of Customer Data, Order Data, and Reviews enables businesses to make informed decisions, fostering a data-driven approach to success in the dynamic world of e-commerce.

#Question 1 : What payment methods are most widely used in e-commerce?

```python
plt.figure(figsize=(8, 8))

pm = sns.countplot(data=df, x="payment_type")
plt.xlabel("Payment Method")
plt.ylabel("")
plt.title("Chart of Payment Method")

for p in pm.patches:
    pm.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

From the data we can know that most of the customer used credit_card as their payment method. This could be supported by the ease of access offered and established between e-commerce parties and credit card service providers. Apart from that, there is the possibility that there will be interesting promotions such as discounts or cashback provided by e-commerce for customers who use credit card payment methods.

#RFM

```python
rfm =df.copy()
```

```python
rfm['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
```

```python
import datetime as dt
current_date = rfm['order_purchase_timestamp'].max() + dt.timedelta(days = 1)

print(f"Current date is: {current_date}")
```

```python
recency_df = rfm.groupby(['customer_unique_id'], as_index = False)['order_purchase_timestamp'].max()

recency_df.head()
```

```python
recency_df['recency'] = recency_df['order_purchase_timestamp'].apply(lambda x: (current_date - x).days)

recency_df.head()
```

```python
frequency_df = pd.DataFrame(rfm.groupby(["customer_unique_id"], as_index = False).agg({"order_id":"nunique"}))

frequency_df.rename(columns={"order_id":"frequency"}, inplace=True)

frequency_df.head()
```

```python
monetary_df = rfm.groupby('customer_unique_id', as_index=False)['payment_value'].sum()

monetary_df.rename(columns={"payment_value":"monetary"}, inplace=True)

monetary_df.head()
```

```python
rfm_df = recency_df.merge(frequency_df, on='customer_unique_id')
rfm_df = rfm_df.merge(monetary_df, on='customer_unique_id')

rfm_df.head()
```

```python
rfm_df.drop(columns='order_purchase_timestamp', inplace = True)

#Show results
rfm_df.head()
```

```python
plt.figure(figsize=(12, 10))

#Recency
plt.subplot(3, 1, 1)
sns.distplot(rfm_df['recency'])
plt.title('Recency distribution')
plt.xlabel(None)

#Frequency
plt.subplot(3, 1, 2)
sns.distplot(rfm_df['frequency'])
plt.title('Frequency distribution')
plt.xlabel(None)

#Monetary
plt.subplot(3, 1, 3)
sns.distplot(rfm_df['monetary'])
plt.title('Monetary distribution')
plt.xlabel(None)

plt.tight_layout(pad = 4)
```

```python
features = ['recency', 'frequency', 'monetary']

n = 1

plt.figure(figsize = (10, 8))

for feature in features:
    plt.subplot(3,1,n)
    sns.boxplot(rfm_df[feature], palette = 'pastel').set(xlabel = None)
    plt.title(f'{feature} boxplot')
    n = n + 1

plt.tight_layout(pad = 4)
```

```python
def remove_outlier(input_df, colum_name):
    q1 = input_df[colum_name].quantile(0.05)
    q3 = input_df[colum_name].quantile(0.95)
    iqr = q3-q1
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    index_outliers = input_df.loc[(input_df[colum_name] < fence_low) | (input_df[colum_name] > fence_high)].index
    input_df = pd.DataFrame(input_df.drop(index_outliers.to_list(), axis=0, inplace=True))
    return input_df
```

```python
for i in ["recency", "monetary"]:
    remove_outlier(rfm_df, i)
```

```python
rfm_df["recency_score"] = pd.qcut(rfm_df['recency'], 5, labels=[5, 4, 3, 2, 1])
rfm_df["monetary_score"] = pd.qcut(rfm_df['monetary'], 5, labels=[1, 2, 3, 4, 5])
```

```python
rfm_df['frequency'].value_counts(normalize=True)*100
```

```python
#Creating a function to parse values
def parse_values(x):
    switch = {
        1: 1,
        2: 2,
        3: 3,
        4: 4
    }
    return switch.get(x, 5)
```

```python
#Apply function to create frequency_score column
rfm_df['frequency_score'] = rfm_df['frequency'].apply(parse_values)
```

```python
#Concatenate all scores into one
rfm_df['rfm_score'] = rfm_df.recency_score.astype(str)+ rfm_df.frequency_score.astype(str) + rfm_df.monetary_score.astype(str)

#Show results
rfm_df.head()
```

```python
segments = {
    r'111|112|121|131|141|151': 'Lost customers',
    r'332|322|233|232|223|222|132|123|122|212|211': 'Hibernating customers',
    r'155|154|144|214|215|115|114|113': 'Cannot Lose Them',
    r'255|254|245|244|253|252|243|242|235|234|225|224|153|152|145|143|142|135|134|133|125|124': 'At Risk',
    r'331|321|312|221|213|231|241|251': 'About To Sleep',
    r'535|534|443|434|343|334|325|324': 'Need Attention',
    r'525|524|523|522|521|515|514|513|425|424|413|414|415|315|314|313': 'Promising',
    r'512|511|422|421|412|411|311': 'New Customers',
    r'553|551|552|541|542|533|532|531|452|451|442|441|431|453|433|432|423|353|352|351|342|341|333|323': 'Potential Loyalist',
    r'543|444|435|355|354|345|344|335': 'Loyal',
    r'555|554|544|545|454|455|445': 'Champions'
}
```

```python
#Creating segment column
rfm_df['segment'] = rfm_df['rfm_score'].replace(segments, regex=True)

#Show results
rfm_df.head()
```

```python
#Calculate some calculations grouping each segment
rfm_statistics = rfm_df[["segment", "recency", "frequency", "monetary"]].groupby("segment").agg(['mean','median', 'min', 'max', 'count'])

#Show results
rfm_statistics
```

```python
#Creating new column with percentage of customers by segment
rfm_statistics['ratio'] = (100 * rfm_statistics['recency']["count"]/rfm_statistics['recency']["count"].sum()).round(2)

#Show results
rfm_statistics
```
